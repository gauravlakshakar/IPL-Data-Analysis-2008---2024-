```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
```

```python
# Data Cleaning: Standardize team names
team_name_mapping = {
    "Delhi Daredevils": "Delhi Capitals",
    "Rising Pune Supergiant": "Rising Pune Supergiants"
}
df['batting_team'] = df['batting_team'].replace(team_name_mapping)
df['bowling_team'] = df['bowling_team'].replace(team_name_mapping)
df
```

```python
# Exploration: Top run-scorers and wicket-takers
top_batters = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_bowlers = df[df['is_wicket'] == 1].groupby('bowler')['is_wicket'].count().sort_values(ascending=False).head(10)
print("Top Batters:\n", top_batters)
print("Top Bowlers:\n", top_bowlers)
```

```python
# Excel Dashboard Preparation
pivot = df.pivot_table(index='over', columns='batting_team', values='total_runs', aggfunc='sum')
pivot
```

```python
# Statistical Testing: Does winning the toss increase match win probability?
toss_win_vs_match_win = pd.crosstab(df['batting_team'], df['is_wicket'])
chi2, p, _, _ = chi2_contingency(toss_win_vs_match_win)
print("Chi-square Test P-Value:", p)
```

```python
# Predictive Modeling
features = df[['inning', 'over', 'batsman_runs', 'total_runs']]
target = df['is_wicket']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))

# Insights & Recommendations
print("Teams winning toss are", "more" if p < 0.05 else "not significantly more", "likely to win the match.")
```
